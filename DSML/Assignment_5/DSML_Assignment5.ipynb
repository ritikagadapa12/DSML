```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/car_crashes.csv")
data.head()
```

```python
plt.scatter(data.speeding,data.alcohol,color = "black",alpha = 0.8)
plt.xlabel("Speeding")
plt.ylabel("Alcohol")
plt.title("Scatter plot",fontsize= 15)
```

```python
plt.bar(data.no_previous,data.not_distracted,color = "r",alpha = 0.5)
plt.xlabel("No_previous")
plt.ylabel("Not_distracted")
plt.title("Bar graph",fontsize= 15)
```

```python
plt.hist(data.ins_losses,bins=30,color="blue",edgecolor="black")
plt.title("Histogram",fontsize= 15)
```

```python
plt.stem(data.alcohol,data.speeding ,linefmt = ":", markerfmt = "b", bottom = 200, basefmt = "r")
plt.title("Stem plot",fontsize= 15)
```
